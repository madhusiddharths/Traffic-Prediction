```python
with open('d12_text_station_5min_2024_01_01.txt/d12_text_station_5min_2024_01_01.txt', 'r') as f:
    data = f.read()
regex = re.compile('(?<date>\d\d\/\d\d\/[0-9]{4})\s(?<time>\d\d\:\d\d\:\d\d)\,(?<station>[0-9]+)\,(?<district>12)\,(?<freeway>[0-9]+)\,(?<direction>[NSWE])\,(?<lane_type>(CD)|(CH)|(FF)|(FR)|(HV)|(ML)|(OR))\,(?<length>[\d.]*)\,(?<samples>[\d]+)\,(?<percentage>[\d]+)\,(?<flow>[\d]+|,,,\d+)')
matches = regex.finditer(data)
df = pd.DataFrame([match.groupdict() for match in matches])
df.head()
```

```python
len(df[df['total'].isna()])
```

```python
time_range = pd.timedelta_range(start='00:00:00', end='23:55:00', freq='5min')
time = []
for i in time_range:
    time.append(str(i).split(' ')[-1])
time[:10]
```

```python
# function to impute null values - see below for refined version
from sklearn.impute import KNNImputer
for i in time:
    df_temp = df[df['Sample Time'].str.contains(i)]
    df_temp.drop(['Sample Time'], axis = 1, inplace = True)
    index_vals = df_temp.index
    imputer = KNNImputer(n_neighbors=3)
    df_temp = imputer.fit_transform(df_temp)
    for j in range(len(index_vals)):
        df.iloc[index_vals[j]]['total'] = df_temp[j][3]
df.info()
```

```python
for i in time:
    # Create a boolean mask to select rows based on the condition
    mask = df['Sample Time'].str.contains(i)
    # Use loc to modify the original DataFrame directly
    df.loc[mask, 'total'] = KNNImputer(n_neighbors=3).fit_transform(df[mask].drop(['Sample Time'], axis=1))[:, 3]
df.info()
```

```python
df['total'] = np.round(df['total'], decimals=0)
df.drop(df.columns[1:4], axis = 1, inplace = True)
df.head()
```

```python
df['Sample Time'] = pd.to_datetime(df['Sample Time'])
df.drop(df[df['Sample Time'] > pd.to_datetime('2023-12-31 23:59:00')].index, inplace = True)
df.to_csv('2023_cleaned.csv')
```

```python
df = pd.read_excel('2023/pems_output.xlsx')
x = np.arange(1,82)
for i in range(len(x)):
    path = '2023/pems_output (' + str(x[i]) + ').xlsx'
    df1 = pd.read_excel(path)
    df = pd.concat([df,df1])
df.info()
```

```python
df['day_of_week'] = df['Sample Time'].apply(lambda x:x.weekday()+1)
dummies = pd.get_dummies(df['day_of_week'])
dummies = dummies.astype(int)
df = pd.concat([df, dummies], axis = 1)
df.head()
```

```python
df['Sample Time'] = pd.to_datetime(df['Sample Time'])
df['hour'] = df['Sample Time'].apply(lambda x:x.hour)
dummy_hour = pd.get_dummies(df['hour'])
dummy_hour = dummy_hour.astype(int)
df = pd.concat([df, dummy_hour], axis = 1)
df.head()
```

```python
df['minute'] = df['Sample Time'].apply(lambda x:x.minute)
dummy_minute = pd.get_dummies(df['minute'])
dummy_minute = dummy_minute.astype(int)
df = pd.concat([df, dummy_minute], axis = 1)
df.head()
```

```python
df.columns = df.columns[:5].tolist() + ['day' + str(x) for x in range(1,8,1)] + ['hour'] + ['hour' + str(x) for x in range(24)] + ['minute'] + ['minute' + str(i) for i in range(0,60,5)]
```

```python
df['total'] = df[df.columns[1:4]].astype(float).sum(1)
```
